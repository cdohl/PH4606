# 2. Harmonic Plane Waves

In the following we will consider a constant speed of sound and start with the description of a plane wave propagation. Planar waves have a constant amplitude and phase perpendicular to the direction of propagation. Thus if the wave moves in the x-plane we have:

$$\ppad{p}{x}=\frac{1}{x^2}\ppad{p}{t}\quad . \tag{2.1}$$

The complex solution to the harmonic wave equation is

$$p=A\,e^{i(\omega t - k x)}+B\,e^{i(\omega + k x)}\quad .\tag{2.2}$$

Using the linear Euler Eq. (1.11) we can obtain the particle velocity in a plane wave as

$$\vb{u}=u\,\va{i}=\left[\frac{A}{\rho_0 c}e^{i(\omega + k x)}-\frac{B}{\rho_0 c}e^{i(\omega - k x)}\right]\va{i}\tag{2.3}$$

Please note that the constants $A$ and $B$, pressures and velocities are now all complex.
We can separate the wave travelling in positive $x$-direction from the the wave travelling in negative $x$-direction and designate them with the subscript $+$ and $-$, respectively.

The wave travelling in positive $x$ is

$$p_+=A e^{i(\omega t - k x)}\tag{2.4}$$ 

while in negative $x$-direction is

$$p_-=B e^{i(\omega t + k x)\tag{2.5}}$$

These two waves can be formally written for the other acoustic variables:

\begin{eqnarray}
u_{\pm}=\pm \frac{p_{\pm}}{\rho_0 c}\tag{2.6}\\
s_{\pm}=\frac{p_{\pm}}{\rho_0 c^2}\tag{2.7}\\
\Phi_{\pm}=-\frac{p_{\pm}}{i\omega\rho_0}\tag{2.8}
\end{eqnarray}


Plane waves travelling in an arbitrary direction can be written as

$$p=A e^{i(\omega t-k_x x- k_y y - k_z z)}\tag{2.9}$$

If we insert Eq. (2.9) into the wave equation, Eq. (1.17) we obtain the condition

$$\left(\frac{\omega}{c}\right)^2 = k_x^2 +k_y^2+k_z^2\tag{2.10}\quad .$$

Now we can introduce the propagation vector, $\vb{k}$ of the wave as

$$\vb{k}\equiv k_x \vb{i} +k_y \vb{j} +k_z \vb{k}\tag{2.11}$$

with

$$\left| \vb{k} \right|=\frac{\omega}{c}\tag{2.12}\quad .$$

Thus using the position vector $\vb{r}=x\vb{i}+y\vb{j}+z\vb{k}$ we can rewrite the plane wave solution, Eq. (2.9), as

$$p=A e^{i(\omega t-\vb{k}\vb{r})}\tag{2.13}\quad .$$

The propagation vector points in the direction of wave propagation which can be simply shown. Surfaces of constant phase are defined by the scalar product $\vb{k}\cdot\vb{r}=\mathrm{const}$. The vector $\nabla (\vb{k}\cdot\vb{r})$ is a vector in the normal direction to the constant phase surface. But $\vb{k}=\nabla (\vb{k}\cdot\vb{r})$, thus $\vb{k}$ is pointing normal to the plane of constant phase and thus in the direction of propagation.

## 2.2 Example

Let a plane wave have a constant phase parallel to the z-axis, thus the wave travel in the xy-plane as sketched in [Fig. 1](#fig1). Formally the surface of constant phase is $\vb{k}\cdot\vb{r}=0$; thus if the phase is independent of $z$, $\vb{k}$ can only have $k_x$ and $k_y$ components. We can thus write Eq. (2.9) as

$$p=A e^{i(\omega t-k_x x- k_y y)}\tag{2.14}$$

and the equation for the surface of constant phase is

$$k_x x + k_y y = \mathrm{const.}\tag{2.15} $$

or

$$y=-\frac{k_x}{k_y}x+\mathrm{const.}\tag{2.16}\quad .$$

Equation (2.16) is line with a slope $-k_x/k_y$. The surfac.
es are thus oriented as sketched in [Fig. 1](#fig1).
