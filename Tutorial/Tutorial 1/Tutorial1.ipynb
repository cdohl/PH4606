## Wave

The wave function gives a mathematical description of a wave. In short, it gives the displacement $y$ of a particle at time $t$ and position $x$:

\begin{equation}
y = A cos(kx \pm \omega t)\quad \tag{1.1}
\end{equation}

where $A$ is the amplitude, $k$ is the wave number and $\omega$ is the angular velocity. The sign is $-$ if the wave propagates in negative direction, and $+$ otherwise. This equation is the solution to the wave equation (1.16) in the lecture note. Try it !!!

### Intuitively, what is $k$ and $\omega$?

$k$ represents the wavelength $\lambda$ of the wave:

\begin{equation}
k = \frac{2 \pi}{\lambda}
\end{equation}

$\omega$ represents the frequency $f$ or the period $T = \frac{1}{f}$ of the wave:

\begin{equation}
\omega = 2\pi f = \frac{2\pi}{T}
\end{equation}

### What about velocity?

When talking about wave, there are two kinds of velocity:

* The velocity of the particle's movement along the wave:
\begin{equation}
v_y = \frac{\partial y}{\partial t} = \omega A sin(kx-\omega t)
\end{equation}

* The velocity of the wave itself:
\begin{equation}
v_{propagate} = \frac{\lambda}{T}
\end{equation}

whereas the second velocity depends on the environment, which, in case of sound, is calculated as:

\begin{equation}
v_{propagate} = c = \frac{\cal B}{\rho_{0}}
\end{equation}

### Now let's plot the wave

First we need to import neccessary libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt  # for plotting
import numpy as np  # for numeric operation
```

We then set the parameters: amplitude $A$, wavelength $\lambda$ and frequency $f$, and from that we calculate wave number $k$ and $\omega$

```python
# Set the parameters
A = 1
wavelength = 1
freq = 1

# Calculate k and omega
k = 2.*np.pi/wavelength
omega = 2.*np.pi*freq
```

Good! Now we left with $x$ and $t$.
First let's create $x$, which will be an array that consists of many elements. Then for each element of $x$, we calculate the corresponding $y$ based on equation (1.1). Each set of $(x,y)$ forms a point on the graph, and connecting those points give us the shape of the wave.
Thus, the more point you have, the smoother the graph look like. Think of it as the 'resolution' of your graph.
For beginning, let's plot the wave at time $t=0$

```python
t = 0
x = np.linspace(0, 2, 20)
y = A * np.cos(k*x - omega*t)
print(x)
# Plot the graph
fig = plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Plotting the wave equation')
```

Looks kinda bad, isn't it? The reason is we have only 20 data points. Let's see what it will become if we increase the number of points to 100.

```python
t = 0
x = np.linspace(0, 2, 100)
y = A * np.cos(k*x - omega*t)

# Plot the graph
fig = plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Plotting the wave equation')
```

As you can see, it gets much better. Now let's plot the wave at a later time $t=0.2$

```python
# The intial wave
t = 0
x = np.linspace(0, 2, 100)
y = A * np.cos(k*x - omega*t)

# The later wave
t1 = 0.2
y1 = A * np.cos(k*x - omega*t1)

# Plot the wave
fig = plt.figure(figsize=(8,6))
plt.plot(x, y, color='blue', label = '$t=0$')
plt.plot(x, y1, color='green', linestyle='--', label = '$t=0.2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='best')
plt.title('Plotting the wave equation')
```
