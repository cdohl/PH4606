# Transmission from one fluid to another: oblique incidence

The boundary is located at $x=0$ while the incident, reflected and transmitted waves make the respective angles $\Theta_i$, $\Theta_r$, and $\Theta_t$, see [Fig. 6.1](#fig6.1).

\begin{equation}
p_i=P_i\,e^{i\left(\omega t - k_1 x \cos \Theta_i - k_1 y \sin \Theta_i\right)}\label{eq:6.1}\tag{6.1}\\
\end{equation}
\begin{equation}
p_r=P_r\,e^{i\left(\omega t + k_1 x \cos \Theta_r - k_1 y \sin \Theta_r\right)}\label{eq:6.2}\tag{6.2}\\
\end{equation}
\begin{equation}
p_t=P_t\,e^{i\left(\omega t - k_2 x \cos \Theta_t - k_2 y \sin \Theta_t\right)}\label{eq:6.3}\tag{6.3}
\end{equation}

Here we use real values for the incident and the reflected angle, while the transmitted angle is allowed to be complex valued.

As the boundary $x=0$ and the pressure is continuous, thus we can write:

\begin{equation}
P_i e^{-i k_1 y \sin \Theta_i} +P_r e^{-i k_1 y \sin \Theta_r}= P_t e^{-i k_2 y \sin \Theta_t}\label{eq:6.4}\tag{6.4}
\end{equation}

Equation (6.4) holds for all $y$, thus we can set $y=0$ and obtain $P_i+P_r=P_t$. Inserting this expression into Eq. (6.4) we obtain

\begin{equation}
P_i e^{-i k_1 y \sin \Theta_i} +P_r e^{-i k_1 y \sin \Theta_r}= P_i e^{-i k_2 y \sin \Theta_t}+ P_r e^{-i k_2 y \sin \Theta_t}\label{eq:6.5}\tag{6.5}\quad .
\end{equation}

Again, as Eq. (6.5) holds for all $y$ the exponents must be the same, which demands

\begin{equation}
k_1 \sin\Theta_i = k_2 \sin\Theta_t\label{eq:6.6}\tag{6.6}
\end{equation}
\begin{equation}
k_1 \sin\Theta_r = k_2 \sin\Theta_t\label{eq:6.7}\tag{6.7}\quad .
\end{equation}

Subtracting Eq. (6.7) from (6.6) leads to the expression that the incident angle is equal to the reflected angle or

\begin{equation}
\sin\Theta_i = \sin\Theta_r\label{eq:6.8}\tag{6.8}\quad .
\end{equation}

While rearranging Eq. (6.6) gives Snell's law

\begin{equation}
\frac{\sin \Theta_i}{c_1}=\frac{\sin \Theta_t}{c_2}\label{eq:6.9}\tag{6.9}\quad .
\end{equation}

Next we want to determine the reflection and transmission coefficients as a function of the incidence angle.

We start wih the second boundary condition, that the normal particle veloicty is continuous at the boundary. This can be written as

\begin{equation}
u_i\cos\Theta_i+u_r\cos\Theta_r=u_t\cos\Theta_t\label{eq:6.10}\tag{6.10}\quad .
\end{equation}

We can replace the velocities with pressures using the specific acoustic impedance (and taking  care of the sign), e.g. $r_1=p_i/u_i$ and $r_1=-p_r/u_r$. Then we obtain 

\begin{equation}
\frac{1}{r_1}\cos\Theta_i-\frac{P_r}{P_i}\frac{\cos\Theta_r}{r_1}=
\frac{P_t}{P_i}\frac{\cos\Theta_t}{r_2}\label{eq:6.11}\tag{6.11}\quad,
\end{equation}
and see that the reflecion coefficient is appearing at the R.H.S. and the transmission coefficient on the L.H.S. of Eq. (6.11):

\begin{equation}
\frac{1}{r_1}\cos\Theta_i-\frac{P_r}{P_i}\frac{\cos\Theta_r}{r_1}=
\frac{P_t}{P_i}\frac{\cos\Theta_t}{r_2}\label{eq:6.12}\tag{6.12}\quad,
\end{equation}

Inserting $R$ and $T$ into Eq. (6.12) we obtain an expression relating the reflection and the transmission coefficient:

\begin{equation}
1-R=\frac{r_1}{r_2}\frac{\cos \Theta_t}{\cos \Theta_i} T\label{eq:6.13}\tag{6.13}
\end{equation}

We can get rid of the transmission coefficient $T$ using the relationship 

\begin{equation}
1+R=T\label{eq:6.14}\tag{6.14}
\end{equation}

and find the relation named *Rayleigh reflection coefficient*:

\begin{equation}
R=\frac{r_2\cos\Theta_i-r_1\cos\cos\Theta_t}{r_2\cos\Theta_i+r_1\cos\Theta_t}=
\frac{\frac{r_2}{\cos \Theta_t}-\frac{r_1}{\cos \Theta_i}}
{\frac{r_2}{\cos \Theta_t}+\frac{r_1}{\cos \Theta_i}}\label{eq:6.15}\tag{6.15}\quad .
\end{equation}

We can use Snell's law to express the cosine of the angle of the transmitted sound $\Theta_t$ as a function of the incidence angle $\Theta_i$ such as

\begin{equation}
\cos \Theta_t= \sqrt{1-\sin^2 \Theta_t}=\sqrt{1-\left(\frac{c_2}{c_1}\right)^2\sin^2\Theta_i}
\label{eq:6.16}\tag{6.16}\quad .
\end{equation}


## Transmission and reflection scenarios

### Scenario 1: $c_1 > c_2$

Snell's law Eq. (6.9) can be rewritten as

\begin{equation}
\sin \Theta_t = \frac{c_2}{c_1}\sin \Theta_i\label{eq:6.17}\tag{6.17}
\end{equation}

We always find a real-valued solution to Eq. (6.17) if $c_2>c_1$. Then the L.H.S. of Eq. (6.17) is smaller than the R.H.S., thus the acoustic wave is refracted towards the surface normal, or in other words $\Theta_t<\Theta_i$.

### Scenario 2: $c_1 < c_2$ and $\Theta_i<\Theta_c$

If $c_1<c_2$ we still find a real-valued solution as long as the incidence angle is $\Theta_i<\Theta_c$, with 

\begin{equation}
\sin \Theta_c=c_1/c_2 \label{eq:6.18}\tag{6.18}
\end{equation}

$\Theta_c$ is the critical angle under which transmission into medium 2 occurs. From Snell's law it follows that $\sin \Theta_t=c_2/c_1 \sin\Theta_i$, thus $\Theta_t$ is always larger than $\Theta_i$ and the beam is refracted away from the normal.

### Scenario 3: $c_1 < c_2$ and $\Theta_i>\Theta_c$

For this scenario the argument of the squareroot in Eq. (6.16) is negative and we obtain a complex result.

\begin{equation}
\cos \Theta_t = -i \sqrt{\left(\frac{c_2}{c_1}\right)^2 \sin^2\Theta_i -1}\label{eq:6.19}\tag{6.19}
\end{equation}

Inserting this expression into the pressure of the transmitted sound, Eq. (6.3), we obtain

\begin{equation}
p_t=P_t e^{-\gamma x} e^{i(\omega t - k_1 y \sin \Theta_i)}\label{eq:6.20}\tag{6.20}
\end{equation}
with 
\begin{equation}
\gamma=k_2 \sqrt{\left(\frac{c_2}{c_1}\right)\sin^2\Theta_i-1} \label{eq:6.21}\tag{6.21}
\end{equation}

Note that the transmitted wave propagates along the interface, in positive y-direction, and is exponentially damped in the second medium. This waves are called *evanescent waves*, they do not transport energy into the second medium.

Let's have a look at the reflection coefficient $R$. It is purely complex and can be written as (after some calculations)

\begin{equation}
R=e^{i\phi}=\cos\phi+i\sin\phi \label{eq:6.22}\tag{6.22}
\end{equation}
with
\begin{equation}
\phi=2 \tan^{-1} \left[\frac{\rho_1}{\rho_2}
\sqrt{\left(\frac{\cos\Theta_c}{\cos\Theta_i}\right)^2-1}\right]
\label{eq:6.23}\tag{6.23}\quad .
\end{equation}

If $\Theta_i\approx\Theta_c$ we can obtain for $\phi=0$ and thus $R=1$; that means the wave is totally reflected with unchanged phase. 

If we have a grazing incidence, that is $\Theta_i \rightarrow \pi/2$ the $\phi\rightarrow\pi$ and the reflection coefficient is $R=-1$, thus the interface resembles a pressure release surface.

## Angle of intromission $\Theta_I$

Looking at the formula for the reflection coefficient Eq. (6.15) we see that the reflection coefficient becomes 0 for real $\theta_t$ and

\begin{equation}
\frac{r_2}{r_1}=\frac{\cos \Theta_t}{\cos \Theta_i}\label{eq:6.24}\tag{6.24}\quad .
\end{equation}

If we combine Eq. (6.24) with Snell's law Eq. (6.9) to eliminate $\Theta_t$ we obtain after some calculations

\begin{equation}
\sin \Theta_I = \sqrt{\frac{\left(\frac{r_2}{r_1}\right)^2 -1}
{\left(\frac{r_2}{r_1}\right)^2-\left(\frac{c_2}{c_1}\right)^2}}=
\sqrt{\frac{1-\left(\frac{r_1}{r_2}\right)^2}{1-\left(\frac{\rho_1}{\rho_2}\right)^2}}\label{eq:6.25}\tag{6.25}
\end{equation}

This defines the angle of intromission. It can only exists if

1.) $r_1 < r_2$ and $c_2 < c_1$  <p>
2.) $r_1 > r_2$ and $c_2 > c_1$

For the second case there is a critical angle. The angle of intromission is smaller than the critical angle.
