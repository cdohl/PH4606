# Acoustic Variables

Acoustic waves travel in solids, liquids and gases. In this course we focus on acoustic waves in liquids and in gases. We do not decribe the motion of each individual molecule but take averages over a number of molecules. This averaged volume is called a fluid particle/element. Because of the smallness of the molecules and their large number continuum description holds for all the this course, yet depending on the kind of material and situation the contiuum assumption may be violated. 
The fluid properties in which acoustic waves propagate described though field variables, i.e. $f(\vb{r},t)$ where $r$ is the equilibrium position of a fluid element and $t$ the time.
Now we introduce the variables to describe the propagation of acoustic waves.
 
#### Equilibrium position of a fluid element

$\va{r}=x\,\va{i}+y\,\va{j}+z\,\va{k}$  where $\va{i}, \va{j}, \va{k}$ are the unit vectors.

#### Particle displacement
$\va{\xi}=\xi_x\,\va{i}+\xi_y\,\va{j}+\xi_z\,\va{k}$ of a fluid particle from its equilibrium position

#### Particle velocity
$\va{u}=\fud{\va{\xi}}{t}=u_x\,\va{i}+u_y\,\va{j}+u_z\,\va{k}$ of a fluid element

#### Density and condensation
$\rho$ is the instanteneous density at position $\va{r}$ <p>
$\rho_0$ is the equilibrium density at position  $\va{r}$ <p>
$s=(\rho-\rho_0)/\rho_0$ condenstation at position  $\va{r}$, this is a normalized amplitude of the density<p>
$\rho-\rho_0=s\cdot\rho_0$ is the acoustic density

#### Pressure
$\cal{P}$ is the instantaneous pressure at $\va{r}$ <p>
$\cal{P}_0$ is the equilibrium pressure at  $\va{r}$ <p>
$p=\cal{P}-\cal{P}_0$ is the acoustic pressure at  $\va{r}$

#### Speed of sound
$c$ is the thermodynamic speed of sound of the fluid

#### Velocity potential
$\Phi$ is the velocity potential at $\va{r}$ with $\va{u}=\nabla \Phi$

#### Temperature
$T$ is the temperature in Kelvins (K) <p>
$T_C$ is the temperature in degrees Celcius (or centigrades) $^\circ$C, thus $T=T_C+273.15\,$K



# Derivation of the Equation of State (EOS)

The ideal gas law comes in many flavors, such as ${\cal P} V = n {\cal R} T$ where $V$ is the volume of the gas, and $\cal{R}$ is the universal gas constant (${\cal R}\approx8.314\,$J mol$^{−1}$K$^{−1}$). We can rewrite this law with the specific gas constant ${\cal R_s}$ as

\begin{equation}{\cal P}=\rho {\cal R}_s T\quad .\label{eq:1.1}\tag{1.1}
\end{equation}

Consider a thermodynamic process within a container with constant wall temperature, then we can describe the relation between density $\rho$ and pressure $\cal P$ as

\begin{equation}
{\cal \frac{P}{P_0}}=\frac{\rho}{\rho_0}\quad .\label{eq:1.2}\tag{1.2}
\end{equation}

If the process is sufficient fast, no heat transfer will occur, and the process is called adiabatic. Then Eq. (\ref{eq:1.1}) becomes

\begin{equation}
{\cal \frac{P}{P_0}}=\left(\frac{\rho}{\rho_0}\right)^\gamma\label{eq:1.3}\tag{1.3}\quad ,
\end{equation}

where $\gamma$ is the ratio of the specific heats. In acoustics many processes can be described sufficiently well with an adiabatic process, however if dissipation needs to be accounted the heat transfer has to be modelled explicitly.

If the fluid does not follow an ideal gas law we can relate the pressure through a Taylor expansion, i.e.

\begin{equation}
{\cal P} = {\cal P_0} +\left(\pad{{\cal P}}{\rho}\right)_{\rho_0}(\rho-\rho_0)
         +\frac{1}{2}\left(\ppad{{\cal P}}{\rho}\right)_{\rho_0}(\rho-\rho_0)^2+ ...
         \label{eq:1.4}\tag{1.4}
\end{equation}

thus if the variations in the density are small only the linear term $\rho-\rho_0$ needs to be accounted for, and Eq. (\ref{eq:1.4}) becomes

\begin{equation}
\cal P-P_0 \approx B\,\frac{\rho-\rho_0}{\rho_0}\quad\label{eq:1.5}\tag{1.5} ,
\end{equation}

where $\cal B=\rho_0\left(\pad{P}{\rho}\right)$ is the adiabatic bulk modulus. We can write Eq. (\ref{eq:1.5}) using the acoustic pressure and the condensation (normalized density amplitude):

\begin{equation}
p\approx {\cal B}\,s\quad . \label{eq:1.6}\tag{1.6}
\end{equation}

Liquids are commonly modeled as adiabats using a matching $\cal P_0$ and $\gamma$ to first order. These two parameters can be related to the measurable quantities of the fluid. These can be revealed from a Taylor approximation of Eq. (\ref{eq:1.3}) in $s$.

After some calculations (see homework) we obtain the following two relations:

\begin{eqnarray}
\gamma{\cal P_0} &=& {\cal B}\\
\gamma&=& \frac{\rho_0}{\cal B} \left(\pad{\cal B}{\rho}\right)_{\rho_0}+1\label{eq:1.7}
\end{eqnarray}

For most fluids $\gamma$ lies between 4 and 12, while $\cal P_0$ is between $10^3\,$bar and $5\cdot 10^3\,$bar. Note that these are fiticious parameters for liquids, while $\gamma$ and $\cal P_0$ for ideal gases have a physical meaning. 

# Derivation of the Linear Wave Equation

We use the linear Euler equation and the linear conservation of mass equation to derive the acoustic wave equation. 

The conservation of mass states 

\begin{equation}
\pad{\rho}{t} + \nabla \cdot (\rho \vb{u})=0\label{eq:1.8}\tag{1.8}
\end{equation}

We now insert $\rho=\rho_0 (1+s)$ and demand that $\rho_0$ changes very little with time, and for $s\ll 1$ we can approximate Eq. (\ref{eq:1.8}) as

\begin{equation}
\rho_0 \pad{s}{t} + \nabla \cdot (\rho_0 \vb{u})=0\label{eq:1.9}\tag{1.9}
\end{equation}

Now if $\rho_0$ is a weak function of space then we obatin the linear conservation of mass equation:

\begin{equation}
\pad{s}{t} + \nabla \cdot \vb{u}=0\label{eq:1.10}\tag{1.10}\quad .
\end{equation}

The linear Euler equation without derivation is 

\begin{equation}
\rho_0\pad{\vb{u}}{t}=-\nabla p\label{eq:1.11}\tag{1.11}\quad .
\end{equation}

The linear wave equation can be derived from Eqs. (\ref{eq:1.10}) and (\ref{eq:1.11}). 

We first take the divergence of the linear Euler Eq. (\ref{eq:1.11}):

\begin{equation}
\nabla \cdot \left(\rho_0\pad{\vb{u}}{t}\right)=-\nabla^2 p\label{eq:1.12}\tag{1.12}\quad .
\end{equation}

Then we take the time derivative of the linear conservation of mass Eq. (\ref{eq:1.10}):

\begin{equation}
\rho_0\ppad{s}{t} + \nabla \cdot \left(\rho_0 \pad{\vb{u}}{t} \right)=0\label{eq:1.13}\tag{1.13}\quad ,
\end{equation}

where we have used the that $\rho_0$ is only weakly dependent on $t$, $\pad{\rho_0}{t}\ll\pad{s}{t}$, and time and space derivations can be interchanged.

We see that the L.H.S of Eq. (\ref{eq:1.12}) is found in Eq. (\ref{eq:1.13}). We can thus write Eq. (\ref{eq:1.13}) as:

\begin{equation}
\nabla^2 p = \rho_0 \ppad{s}{t}\label{eq:1.14}\tag{1.14}
\end{equation}

Because $s=p/\cal B$ and $\cal B$ being a weakly dependent on time, we obtain

\begin{equation}
\nabla^2 p = \frac{\rho_0}{\cal B} \ppad{s}{t}\label{eq:1.15}\tag{1.15}\quad ,
\end{equation}

which is a linear lossless wave equation with a speed of sound 

\begin{equation}
c^2= \frac{\cal B}{\rho_0}\label{eq:1.16}\tag{1.16}\quad .
\end{equation}

Inering Eq. (\ref{eq:1.16}) into Eq. (\ref{eq:1.15}) we obtain the linear wave equation:

\begin{equation}
\nabla^2 p -\frac{1}{c^2}\ppad{p}{t} = 0\label{eq:1.17}\tag{1.17}\quad .
\end{equation}

We can show that condensation fulfills the wave equation, too: using Eq. (\ref{eq:1.6}) we obtain an equation fro the pressure from Eq. (\ref{eq:1.16}) as

\begin{equation}
p=\rho_0 c^2 s\label{eq:1.18}\tag{1.18}\quad.
\end{equation}

Inserting Eq. (\ref{eq:1.18}) into Eq. (\ref{eq:1.17}) and demanding that $\rho_0$ and $c$ have a weak dependency on space gives a wave equation for $s$.

# Velocity Potential

The curl of a gradient of any function is 0, i.e. $\nabla\times\nabla f(\va{r})=0$. If we apply a curl on both sides of Eq. (\ref{eq:1.11}), thus:

\begin{eqnarray}
\nabla \times \left(\rho_0\pad{\vb{u}}{t}\right)&=&\nabla \times \left(-\nabla p\right)\\
\rho_0\pad{}{t}\left(\nabla\times\vb{u}\right)&=&0\label{eq:1.19}
\end{eqnarray}

Equation (\ref{eq:1.19}) is valid for $\rho_0$ being a weak function of space. Then however, we see that the curl of the velocity field must be zero. Any vector field with zero curl can be written as a gradient of a scalar field. Thus we can derive the particle velocity field $\vb{u}(\vb{r},t)$ from a scalar potential field, the so-called velocity potential $\Phi$:

\begin{equation}
\vb{u}=\nabla \Phi\label{eq:1.20}\tag{1.20}
\end{equation}

This simplifies the problem a lot as we now need only to solve for a scalar field instead of a vector field. We can show that the the wave equations applies to the velocity field, too. Therefore, we substitute Eq. (\ref{eq:1.20}) into the linear Euler Eq. (\ref{eq:1.11}) and obtain
    
\begin{equation}
\nabla\left(\rho_0\pad{\vb{u}}{t}+p\right)=0\label{eq:1.21}\tag{1.21}
\end{equation}

Thus the expression in the bracket of Eq. (\ref{eq:1.21}) is constant. We can arbitrarily set this constant to 0 and obtain the relationship between pressure and velocity potential

\begin{equation}
p=-\rho_0\pad{\Phi}{t}\label{eq:1.22}\tag{1.22}\quad .
\end{equation}

It can be easily shown that $\Phi$ satisfies the wave equation, too.

# Speed of Sound in Fluids

By combining our previous Eq. (\ref{eq:1.5}) 

\begin{equation*}
\cal P-P_0 \approx B\,\frac{\rho-\rho_0}{\rho_0}
\end{equation*}

and Eq. (1.16) 

\begin{equation*}
c^2= \frac{\cal B}{\rho_0}
\end{equation*}

we obtain an expression for the speed of sound

\begin{equation}
c^2\approx \frac{\cal P-P_0}{\rho-\rho_0}=\pad{\cal P}{\rho}\label{eq:1.23}\tag{1.23}\quad .
\end{equation}

Using the adiabatic relationship, Eq. (\ref{eq:1.3}), between pressure and density we can take the derivative of Eq. (\ref{eq:1.23}) and obtain

\begin{equation*}
\pad{\cal P}{\rho}=\gamma \frac{\cal P}{\rho}
\end{equation*}

let's choose $\cal P=P_0$  and $\rho=\rho_0$, that is the pressure and denisty at some special condition, e.g. $T=0^\circ$C and $\cal P = 1\,$atm.
Thus the prediction of the speed of sound is

\begin{equation*}
c^2=\gamma \frac{\cal P_0}{\rho_0}\approx \frac{1.402 1.01325\times 10^5 \,\textrm{Pa}}{1.293\,\textrm{kg/m}^3}=(331.5\,\textrm{m/s})^2\quad\,
\end{equation*}


which is in excellent agreement with measurements (please check the  [table](https://en.wikipedia.org/wiki/Speed_of_sound#Tables)).

For most gases the ratio $\cal P_0 /\rho_0$ for a constant temperature is independent of the pressure, thus the speed of sound is mostly a function of the temperature.

Using the equation for a perfect gas ${\cal P}=\rho {\cal R}_s T$, see Eq. (\ref{eq:1.1}), Eq. (\ref{eq:1.23}) becomes 

\begin{equation}
c^2=\gamma {\cal R}_s T\label{eq:1.24}\tag{1.24}\quad .
\end{equation}

This expression can be conveniently written in $T_C$, temperature measure in degree Celsius:

\begin{equation}
c=c_0 \sqrt{1+T_C/273}\label{eq:1.25}\tag{1.25}\quad ,
\end{equation}

where $c_0$ is the speed of sound at $T=0^\circ$C.

Below figure depicts a comparison between measurements and Eq. (\ref{eq:1.25}).

```python
%matplotlib inline
import numpy#array operations
import matplotlib.pyplot as plt #plotting
#taken from https://en.wikipedia.org/wiki/Speed_of_sound#Tables
data=numpy.array([[35,351.88],[30,349.02],[25,346.13],[20,343.21],[15,340.27],\
                  [10,337.31],[5,334.32],[0,331.30],[-5,328.25],[-10,325.18],\
                  [-15,322.07],[-20,318.94],[-25,315.77]])
TC=numpy.arange(-50,50,1)
c0=331.5
plt.plot(TC,c0*numpy.sqrt(1.+TC/273.),data[:,0],data[:,1],'o')
plt.xlabel('T ($^\circ$C)')
plt.ylabel('c (m/s)');
```

## Simulation of a 1-dimensional wave

The program below allows to solve the 1-dimensional wave equation. To execute it, you need to place the cursor in each cell and press SHIFT + RETURN. The program consists of code and an interface, thus please press SHIFT + RETURN in each cell. Then you should see the sliders and buttons. To start a simple case just press the button labelled "Start Simulation". Please start playing and exploring the program, we'll work with it in the class, too.

```python
%matplotlib notebook 
import math as m
import numpy #array operations
import matplotlib.pyplot as plt #plotting
from IPython import display #for continous display
from ipywidgets import widgets #for the widgets

nimg=0

def plotwave(u,time,xx):
    plt.figure(1)
    plt.clf()
    plt.plot(xx,u)
    plt.gca().set_ylim([-2,2])
    plt.text(0.1,1.8,"time {0:.5f}".format(time)) #annotate the time
    display.clear_output(wait=True)
    display.display(plt.gcf())
        
def solvewave(b):
    tabs.visible=False
    #computational domain
    nx = 381
    size=2. #size of the domain
    #parameters of the wave
    c = 5. #speed of sound
    l=w_wavelength.value #wavelength
    nu=c/l #frequency
    omega=nu*2.*m.pi #angular frequency
    duration=w_sourceduration.value/nu #duration of source
    
    #further variables
    dx = size/(nx-1)
    CFL=0.1 #CFL number <1
    dt = CFL*dx/c 
    nt=int(w_simduration.value/dt) #number of time steps
    
    x_axis=numpy.arange(nx,dtype='float')*dx-1.
    if w_position.value=='Left':
        sourcepos=0
    else:
        sourcepos=int(nx/2)
    
    #every xx times over the total nt timesteps an output should be generated 
    output=map(int,list(numpy.linspace(1,nt,int(nt/50))))

    u  = numpy.zeros(nx) #pressure at t
    un = numpy.zeros(nx) #pressure at t-dt
    unn= numpy.zeros(nx) #pressure at t-2*dt
    C=c*c*dt*dt/dx/dx
    
    #Assign initial conditions, for a sine wave
    if w_source_type.value=="Time Dependent":
        un[sourcepos]=0.    #amplitude is sin(omega*t) with t=0
        unn[sourcepos]=1.   #velocity  is cos(omega*t) with t=0
    else: #or set it as an initial value
        for xx in range(nx):
            x=float(xx)*dx-size/2. #0 at the center
            unn[xx]=(m.cos(2.*m.pi/l*x)**2)*float(abs(x)<(c*duration/2.))
        for xx in range(1,nx-1): #calculate t=0 time step
            un[xx] = unn[xx] - 0.5*C*(unn[xx+1]-2.*unn[xx]+unn[xx-1])
    
    plt.figure(1, figsize=(8, 4), dpi=300)

    for n in range(nt+1): ##loop across number of time steps
        #this line computes the finite differences of the wave equation
        u[1:-1]=2.*un[1:-1]-unn[1:-1]+C*(un[:-2]+un[2:]-2.*un[1:-1])
 
        #Boundary conditions
        if w_boundary_r.value=='Open':
            u[-1] = un[-1]-dt*c/dx*(un[-1]-un[-2]) 
        else:
            u[-1] = u[-2]

        if w_boundary_l.value=='Open':
            u[0] = un[0]+dt*c/dx*(un[1]-un[0])
        else:
            u[0] = u[1]
            
        #pressure source
        if w_source_type.value=="Time Dependent":
            if float(n*dt<duration):
                u[sourcepos]=m.sin(omega*n*dt)**2*float(n*dt<duration)
                
        #save values for the time derivative 
        unn=un.copy() #n-1 time step
        un=u.copy()   #n time step
        
        if (n in output):
            plotwave(u,n*dt,x_axis)
            
    #and plot the last figure    
    plotwave(u,n*dt,x_axis)
    tabs.visible=True
```

```python
#setup the graphical interface 
w_wavelength=widgets.FloatSlider(description="Wavelength",value=.5,min=0.1,max=1)
w_sourceduration=widgets.FloatSlider(description="Emission Duration in Periods",value=.5,min=0.5,max=10)
w_simduration=widgets.FloatSlider(description="Simulation Duration",value=.2,min=0.1,max=5)
w_position=widgets.RadioButtons(description="Source Position",options=["Left", "Center"],value="Center")
w_start=widgets.Button(description="Start Simulation")
w_start.on_click(solvewave)
w_boundary_l=widgets.RadioButtons(description="Leftt Boundary",
                                  options=["Reflective", "Open"],value="Reflective")
w_boundary_r=widgets.RadioButtons(description="Right Boundary",
                                  options=["Reflective", "Open"],value="Reflective")
w_source_type=widgets.RadioButtons(description="Type of Source",
                                   options=["Time Dependent", "Inital Condition"],value="Time Dependent")

page1=widgets.Box([w_wavelength,w_sourceduration,w_position,w_simduration])
page2=widgets.Box([w_source_type])
page3=widgets.Box([w_boundary_l,w_boundary_r])
tabs = widgets.Tab(children=[page1, page2, page3])
tabs.set_title(0, 'Wave')
tabs.set_title(1, 'Source')
tabs.set_title(2, 'Boundaries')
display.display(tabs)

display.display(w_start)
```
